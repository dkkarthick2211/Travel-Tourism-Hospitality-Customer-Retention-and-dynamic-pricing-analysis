```python
import pandas as pd                #booking behaviour
df=pd.read_csv("hotel_booking_cleaned.csv")
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['lead_time'], bins=30)
plt.show()                           
```

```python
df.info()
```

```python
sns.countplot(x='customer_type', data=df)     #most common customer type
plt.show()
```

```python
sns.boxplot(x='reservation_status', y='average_daily_rate', data=df)     #higher room price leads to cancellation
plt.show()
```

```python
sns.boxplot(x='reservation_status', y='lead_time', data=df)      #book earliercancel more often
plt.show()
```

```python
pd.crosstab(df['customer_type'], df['reservation_status'])        #group has the highest cancellation rate
```

```python
corr = df.corr(numeric_only=True)       #factor affecting cancellation
sns.heatmap(corr, annot=True)
plt.show()
```

```python
sns.countplot(                   #peakseason,off season,demand trends
    x='Month',
    data=df,
    order=[
        'January','February','March','April',
        'May','June','July','August',
        'September','October','November','December'
    ]
)
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_average_daily_rate = df.groupby('Month')['average_daily_rate'].mean()     #months have the highest room prices
monthly_average_daily_rate.plot(kind='bar')
plt.show()
```

```python
sns.countplot(                #factor affecting cancellations
    x='deposit_type',
    hue='reservation_status',
    data=df
)
plt.show()
```

```python
sns.boxplot(
    x='reservation_status',
    y='previous_cancellations',
    data=df
)
plt.show()
```
